### Load and Resize Image

In this step, we load the input image and resize it to a predefined target size. Resizing the image ensures that it has consistent dimensions and is suitable for further preprocessing steps.

```python
import cv2
import matplotlib.pyplot as plt

# Load input image
input_image = cv2.imread("..\\data\\sign_example.jpg")

# Convert BGR to RGB
input_image_rgb = input_image#cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)

# Define target size for resizing
target_size = (224, 224)

# Resize the image
resized_image = resize_image(input_image_rgb, target_size)

# Display resized image
plt.imshow(resized_image)
plt.title("Resized Image")
plt.axis('off')
plt.show()
```

### Denoise Image

Denoising the image helps to remove noise and unwanted artifacts, resulting in a smoother image. This step improves the quality of the image and prepares it for subsequent processing steps.

```python
# Step 2: Denoise the image
denoised_image = denoise_image(resized_image)

# Display denoised image
plt.imshow(denoised_image)
plt.title("Denoised Image")
plt.axis('off')
plt.show()
```

### Edge Detection and Processing

This function applies edge detection using the Canny edge detector followed by morphological processing. Here's what happens in each step:

1. **Canny Edge Detection**:
   - The Canny edge detector is applied to the input image to detect edges.
   - This detector uses gradient information to find the intensity gradients of the image and then marks the points where the gradient is higher than the upper threshold as edge pixels.

2. **Morphological Processing (Closing)**:
   - After the edges are detected, a morphological operation called closing is performed.
   - Closing is a combination of dilation followed by erosion. It's commonly used to close small gaps or holes in the edges.
   - A structuring element, represented by a 3x3 array of ones, is used for the operation. This structuring element defines the neighborhood considered around each pixel during the operation.
   - The closing operation helps to connect nearby edge pixels and fill in small gaps, resulting in smoother and more continuous edges.


```python
# Step 3: Apply adaptive thresholding
edges_image = edge_detection(denoised_image)

# Display thresholded image
plt.imshow(edges_image, cmap='gray')
plt.title("Edges Image")
plt.axis('off')
plt.show()
```

### Normalize Image

Normalizing the pixel values of the image ensures that they are within a consistent range, typically between 0 and 1. This step is important for standardizing the input to deep learning models and improving their convergence during training.

```python
# Step 5: Normalize pixel values
normalized_image = normalize_image(edges_image)

# Display normalized image
plt.imshow(normalized_image, cmap='gray')
plt.title("Normalized Image")
plt.axis('off')
plt.show()
```
